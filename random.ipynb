```python
import numpy as np
np.random.seed(0)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# random.rand() uniform distribution in (0, 1)
a = np.random.rand(1, 3000)
sns.distplot(a, kde=False)
```

```python
# random.randn() standard normal distribution
a = np.random.randn(1, 3000)
sns.distplot(a)
```

```python
# random.randint(start, stop, size)
a = np.random.randint(2, 10, size=(5, 5))
a
```

```python
# random.normal(mean, std, size)
a = np.random.normal(loc=10, scale=5, size=3000)
sns.distplot(a)
```

```python
# random.bionomial(n, p, size)
a = np.random.binomial(100, 0.4, size=(1, 3000))
sns.distplot(a)
```

```python
# shuffle, only applies to the 1st axis
# the original array is changed
a = np.arange(12)
print(a)
np.random.shuffle(a)
print(a)
b = np.arange(12).reshape(4, 3)
print(b)
np.random.shuffle(b)
print(b)
```

```python
# permutation, only applies to the 1st axis
# returns a new array, the original array is not changed
a = np.arange(12)
b = np.arange(12).reshape(4, 3)
c = np.random.permutation(a)
d = np.random.permutation(b)
print(a)
print(b)
print(c)
print(d)
```
